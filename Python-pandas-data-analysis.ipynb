# Merging data from multiple sources 

```python
urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/locations.csv',
            'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location == 'Italy']
```

```python
covid_df['location'] = "Italy"
```

```python
covid_df
```

```python
# locations_df into covid_df using the .merge 
merged_df = covid_df.merge(locations_df, on= 'location')
```

```python
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases* 1e6/merged_df.population
```

```python
merged_df['tests_per_million'] = merged_df.total_cases* 1e6/merged_df.population
```

```python
merged_df['deaths_per_million']= merged_df.total_cases * 1e6/merged_df.population
```

```python
merged_df
```
